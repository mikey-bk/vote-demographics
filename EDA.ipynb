```python
import census_reports as cr
```

```python
rpt = cr.CensusReports()
```

```python
df=rpt.get_flattened_frame(2016)
```

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
df = df[~pd.isna(df['pvi'])]
```

```python
df
```

```python
df['pvi'].plot.hist()
```

```python
df['pvi'].plot.box()
```

```python
df.columns
```

```python
df.loc[:,'Density']=df['Total Population'] / df['ALAND']
```

```python
data = df[['HIGH_SCHOOL_AND_HIGHER_PCT', 'BACHELOR_AND_HIGHER_PCT', 'RACE_WHITE_NON_HISPANIC_PCT', 'RACE_BLACK_NON_HISPANIC_PCT', 'RACE_NATIVE_AM_NON_HISPANIC_PCT', 'RACE_ASIAN_NON_HISPANIC_PCT', 'HISPANIC_OR_LATINO_PCT', 'MARRIED_COUPLE_FAMILIES_PCT', 'NON_FAMILY_HOUSEHOLDS_PCT', 'INCOME_HOUSEHOLDS_BELOW_35K_PCT', 'INCOME_HOUSEHOLDS_ABOVE_200K_PCT', 'HOUSEHOLD_MEDIAN_INCOME', 'TOTAL_POP_18+_PCT', 'TOTAL_POP_65+_PCT', 'Density', 'pvi']]
```

```python
data.loc[:,'AVG_HOUSEHOLD_SIZE'] = df['Total Population'] / df['NUM_HOUSEHOLDS'] 
```

```python
data
```

```python
data['AVG_HOUSEHOLD_SIZE'].plot.hist()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data[['AVG_HOUSEHOLD_SIZE', 'pvi']])
```
